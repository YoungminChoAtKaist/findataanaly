```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.optimize import Bounds

# Expected Return
e = np.array([0.169976, 0.107028, 0.192165, 0.003749, 0.084538])

# Covariance Matrix(5x5)
V = np.array([
    [0.002204, 0.000322, 0.001247, 0.000414, 0.000943],
    [0.000322, 0.000666, 0.000297, 0.000186, 0.000189],
    [0.001247, 0.000297, 0.002352, 0.000429, 0.000733],
    [0.000414, 0.000186, 0.000429, 0.000755, 0.000334],
    [0.000943, 0.000189, 0.000733, 0.000334, 0.001394]
])

# initial constraints ( sum is 1 )
constraints = [{'type': 'eq', 'fun': lambda x: np.sum(x) - 1}]

bounds = Bounds(0,1)

def portfolio_variance(weights):
    return weights.T @ V @ weights

def min_variance_for_return(target_return):
    def objective(weights):
        return portfolio_variance(weights)

    constraints_return = constraints.copy()
    constraints_return.append({'type': 'eq', 'fun': lambda x: np.dot(x, e) - target_return})

    initial_weights = np.ones(5) / 5

    # run minimizse
    result = minimize(objective, initial_weights, bounds=bounds, constraints=constraints_return)

    return result.x  

# range of mean
target_returns = np.linspace(-0.1, 0.3, 41)

# results
weights_list = []
portfolio_means = []
portfolio_stds = []

pd.options.display.float_format = '{:.4f}'.format

print('Find optimal weights at each level of the expected return')

for target_return in target_returns:
    weights = min_variance_for_return(target_return)
    weights_list.append(weights)
    
    # calc return and st.dev
    mean_return = np.dot(weights, e)
    std_dev = np.sqrt(portfolio_variance(weights))
    
    portfolio_means.append(mean_return)
    portfolio_stds.append(std_dev)

    optimal_weight = pd.DataFrame(weights)
    optimal_weight.columns = [target_return]
    optimal_weight.index = ['AMAT','BA', 'QCOM', 'SBUX', 'APPL']

    display(optimal_weight)

print('Plot the Minimum Variance Portfolios in a mean-variance space and in a mean-standard deviation space')

# Plot the Minimum Variance Portfolio
plt.figure(figsize=(10, 6))
plt.plot(portfolio_stds, portfolio_means, marker='o', linestyle='-', color='b')
plt.xlabel('Standard Deviation (Risk)')
plt.ylabel('Expected Return')
plt.title('Minimum Variance Portfolio in Mean-Standard Deviation Space')
plt.grid(True)
plt.show()

print('Any potential problem in applying this model?')
```
